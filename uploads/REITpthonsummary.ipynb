```python
from openpyxl import load_workbook
import pandas as pd

file_path = 'C:\\Users\\wxy02\\OneDrive\\anaconda_python_code\\REIT Fund performance(AutoRecovered).xlsx'
wb = load_workbook('C:\\Users\\wxy02\\OneDrive\\anaconda_python_code\\REIT Fund performance(AutoRecovered).xlsx', data_only=True)

target_sheets = ["SE warehouse", "Heathrow warehouse", "office"]

def to_number(val):
    if isinstance(val, (int, float)):
        return val
    if isinstance(val, str):
        try:
            return float(val.replace(",", "").replace("£", "").strip())
        except:
            return None
    return None

def find_value_by_labels(ws, possible_labels, max_offset=5):
    labels = [x.strip().lower() for x in possible_labels]

    for row in ws.iter_rows():
        for cell in row:
            if isinstance(cell.value, str) and cell.value.strip().lower() in labels:
                for offset in range(1, max_offset + 1):
                    val = ws.cell(cell.row, cell.column + offset).value
                    num = to_number(val)
                    if num is not None:
                        return num
    return None

summary_rows = []

for sheet in target_sheets:
    ws = wb[sheet]

    purchase_value = find_value_by_labels(ws, ["purchase value"])
    exit_price = find_value_by_labels(ws, ["exit price"])
    Construction_cost = find_value_by_labels(ws, ["Construction cost"])

    summary_rows.append({
        "Sheet": sheet,
        "Purchase Value": purchase_value,
        "Exit Price": exit_price,
        "Construction cost": Construction_cost
    })

pd.options.display.float_format = '{:,.2f}'.format

summary_df = pd.DataFrame(summary_rows)
summary_df
```
